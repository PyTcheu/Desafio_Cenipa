```python
### Removendo registros duplicados e simplificando à abstração de Recomendação Cumprida ou Não Cumprida
```

```python
recomendacao_f = pd.DataFrame()

for i in recomendacao['codigo_ocorrencia4'].unique():
    df_x = recomendacao[recomendacao['codigo_ocorrencia4'] == i]
    df_x['recomendacao_Cumprida'] = df_x[df_x['recomendacao_status'] == 'CUMPRIDA'].shape[0]
    df_x['recomendacao_NaoCumprida'] = df_x[df_x['recomendacao_status'] == 'NÃO CUMPRIDA'].shape[0]
    df_x['recomendacao_Aguardando'] = df_x[df_x['recomendacao_status'] == 'AGUARDANDO RESPOSTA'].shape[0]
    df_x['recomendacao_CumpridaAlt'] = df_x[df_x['recomendacao_status'] == 'CUMPRIDA DE FORMA ALTERNATIVA'].shape[0]
    df_x['recomendacao_***'] = df_x[df_x['recomendacao_status'] == '***'].shape[0]

    recomendacao_f = pd.concat([recomendacao_f, df_x])
    
recomendacao = recomendacao_f.copy()
```

```python
recomendacao = recomendacao.drop_duplicates(subset=['codigo_ocorrencia4'], keep='last')
```

## Conclusões da análise exploratória

- 3041 Incidentes
- 735 Incidentes Graves
- 1752 Acidentes, dos quais:
	- 1338 Sem fatalidades
	- 414 Com fatalidades, das quais:
		- 330 Com 1 ou 2 fatalidades
		- 219 Com 1 fatalidade
		
		- 98% dos casos com 1 ou 2 fatalidades, a aeronave teve os danos no nível DESTRUÍDA ou SUBSTANCIAL.
		- 50% dos casos com 1 ou 2 fatalidades foram de aeronaves com 1 ou 2 assentos.
        
	#### Hipótese: A maioria dos acidentes com fatalidades ocorrem em operações individuais onde há apenas o piloto.

		- ! Há 4 Acidentes com 1 fatalidade que não apresentou NENHUM dano à aeronave
		- ! Há 15 Acidentes com fatalidades que a aeronave apresenta Assentos = 0

- Em 2012 e 2013 foram os recordes de incidentes registrados (650+)
- Em 2016 o número mais baixo (400+)
- A partir de 2016, voltou a subir lentamente numa média de +25 por ano
- Os estados em que mais houveram Ocorrências foram: SP, MG, PR, RJ e RS
- Os estados em que mais houveram ACIDENTES foram: SP, MT, RS, MG, PR
- Há a possibilidade de estes estados serem os mais comuns pelo volume de atividade aérea das regiões, justificados pela proporção populacional das regiões em relação ao País.
